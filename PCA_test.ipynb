```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.rand(1000, 2), columns=list('AB'))
```

```python
df.head()
```

```python
eig_vals, eig_vecs = np.linalg.eig(np.cov(df.T))
```

```python
eig_vals[1] / sum(eig_vals)
```

```python
projected_X = df.dot(eig_vecs.T[1])
projected_X
```

```python
df['y_axis'] = 0.0
df['PCA'] = projected_X
df.head()
```

```python
import seaborn as sns

sns.lmplot('A', 'B', data=df)
```
